### **1.B)**

```python
import pandas as pd

base = "champions.mod.csv"  
df = pd.read_csv(base)

print("Tipos de Variáveis:")
print(df.dtypes)

numerical_data = df.select_dtypes(include=['float64', 'int64'])

def calcular_estatisticas(df):
    estatisticas = {}
    for coluna in df.columns:
        valores = df[coluna].dropna() 
        estatisticas[coluna] = {
            'Média': valores.mean(),
            'Moda': valores.mode().tolist(), 
            'Mediana': valores.median(),
            'Variância': valores.var(),
            'Amplitude': valores.max() - valores.min(),
            'Mínimo': valores.min(),
            'Máximo': valores.max(),
            'Valores Distintos': valores.nunique()
        }
    return pd.DataFrame(estatisticas)

estatisticas_df = calcular_estatisticas(numerical_data)
print(estatisticas_df.T)
```

### **1.C)**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

base = "champions.mod.csv"  
df = pd.read_csv(base)

numerical_data = df.select_dtypes(include=['float64', 'int64'])

for coluna in numerical_data.columns:
    valores = numerical_data[coluna].dropna()

    plt.figure(figsize=(12, 6))
    sns.histplot(valores, kde=True, stat="density", bins=30, color='skyblue', label='Histograma')
    plt.title(f"Distribuição de {coluna}")
    plt.xlabel(coluna)
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

    plt.figure(figsize=(8, 8))
    stats.probplot(valores, dist="norm", plot=plt)
    plt.title(f"Gráfico Q-Q de {coluna}")
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(x=valores, color='orange')
    plt.title(f"Gráfico de Boxplot de {coluna}")
    plt.xlabel(coluna)
    plt.show()

    stat, p_value = stats.shapiro(valores)
    print(f"Shapiro-Wilk Test para {coluna}: Estatística={stat:.4f}, p-valor={p_value:.4f}")
    if p_value > 0.05:
        print(f"-> {coluna} segue uma distribuição normal (p-valor > 0.05)\n")
    else:
        print(f"-> {coluna} NÃO segue uma distribuição normal (p-valor <= 0.05)\n")
```

```python
from sklearn.model_selection import train_test_split

test_data = df[df.isnull().any(axis=1)].copy()

data_without_missing = df.dropna()

train_data, val_data = train_test_split(data_without_missing, test_size=0.2, random_state=42) 

print("Tamanho da base de treino:", len(train_data))
print("Tamanho da base de validação:", len(val_data))
print("Tamanho da base de teste:", len(test_data))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

Identificar colunas categóricas
categorical_cols = ['Range type', 'Resourse type', 'Tags', 'Role', 'Name']

Aplicar One-Hot Encoding
data_encoded = pd.get_dummies(df, columns=categorical_cols, dummy_na=True)

print("Colunas geradas pelo One-Hot Encoding:")
print(data_encoded.columns)

cols_to_check = ['Movement speed'] + [
    col for col in data_encoded.columns 
    if 'Range' in col or 'Resourse' in col
]

print("\nColunas para checagem: " )
print(cols_to_check)

missing_values = data_encoded[cols_to_check].apply(lambda col: (col == 1).sum())


Identificar colunas com valores ausentes
cols_with_missing = [col for col in data_encoded.columns if data_encoded[col].isnull().any()]

nan_cols = [col for col in data_encoded.columns if col.endswith('_nan') and (data_encoded[col] == 1).any()]

Combinar ambas as listas para obter todas as colunas relevantes
cols_to_evaluate = list(set(cols_with_missing + nan_cols))

print("Colunas com valores ausentes:", cols_to_evaluate)

Prossiga com a modelagem
for col in cols_with_missing:
    data_cleaned = data_encoded.dropna(subset=[col])

    X_train = data_cleaned.drop(col, axis=1)
    y_train = data_cleaned[col]

    linear_model = LinearRegression()
    linear_model.fit(X_train, y_train)

    y_pred = linear_model.predict(X_train)

    mse = mean_squared_error(y_train, y_pred)
    r2 = r2_score(y_train, y_pred)

    if mse < 0.1:
        print(f"{mse:.3f}: O MSE é bom.")
    else:
        print(f"{mse:.3f}: O MSE não é ideal.")

    if r2 > 0.7:
        print(f"{r2:.3f}: O R² é bom.")
    else:
        print(f"{r2:.3f}: O R² não é ideal.")

    print("-" * 20)
```

### **2.B)**

A **regressão linear** é o melhor modelo :

- Apresenta um **MSE extremamente baixo (0.000)**, indicando previsões quase perfeitas.
- Tem um **R² igual a 1.000**, o que significa que o modelo explicou 100% da variância nos dados, sugerindo uma ótima correspondência entre os valores preditos e reais.
- Além disso, ela é **simples**, **rápida** e **fácil de interpretar**, tornando-a uma escolha muito eficiente para dados contínuos com relações lineares claras.

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

base = "champions.mod.csv"
df = pd.read_csv(base)

df_encoded = pd.get_dummies(df)

if df_encoded.isnull().any().any():
    df_encoded = df_encoded.fillna(df_encoded.mean())  

iso_forest = IsolationForest(n_estimators=100, contamination=0.1, random_state=42)

iso_forest.fit(df_encoded)
anomalies = iso_forest.predict(df_encoded)

df['Anomaly'] = anomalies

print("Outliers detectados:")
print(df[df['Anomaly'] == -1]) 

if df_encoded.shape[1] >= 2:  
    plt.scatter(df_encoded.iloc[:, 0], df_encoded.iloc[:, 1], c=df['Anomaly'], cmap='coolwarm', edgecolor='k')
    plt.xlabel(df_encoded.columns[0])
    plt.ylabel(df_encoded.columns[1])
    plt.title('Detecção de Outliers com Isolation Forest')
    plt.show()
```

```python
import pandas as pd
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

df = pd.read_csv(base)

X = df.select_dtypes(include=[float, int]).values  


if not np.all(np.isfinite(X)):
    print("Existem valores não finitos nos dados. Realizando o tratamento...")

    import numpy as np
    colunas = np.isnan(X).any(axis=0)  
    if np.any(colunas):
        X[:, colunas] = np.nan_to_num(X[:, colunas], nan=np.nanmean(X[:, colunas]))

    X = np.nan_to_num(X, posinf=0, neginf=0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dendrogram = sch.dendrogram(sch.linkage(X_scaled, method='ward'))
plt.title('Dendrograma - Método Hierárquico')
plt.xlabel('Pontos de dados')
plt.ylabel('Distância')
plt.show()
```

### **3.C)**

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

df = pd.read_csv(base)

colunas_numericas = df.select_dtypes(include=[float, int]).columns
colunas_categoricas = df.select_dtypes(include=[object]).columns

df_encoded = pd.get_dummies(df, columns=colunas_categoricas)

X = df_encoded.values

if df_encoded[colunas_numericas].isnull().any().any():
    print("Existem valores ausentes nos dados numéricos. Realizando o tratamento...")
    imputer = SimpleImputer(strategy='mean')
    X = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
print("Variância explicada por cada componente:", explained_variance_ratio)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), np.cumsum(explained_variance_ratio), marker='o')
plt.title('Variância Explicada Acumulada pelo PCA')
plt.xlabel('Número de Componentes Principais')
plt.ylabel('Variância Explicada Acumulada')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', marker='o', alpha=0.5)
plt.title('Primeiros dois Componentes Principais')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

### **3.D)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import seaborn as sns

df = pd.read_csv(base)

colunas_numericas = df.select_dtypes(include=[float, int]).columns
colunas_categoricas = df.select_dtypes(include=[object]).columns

df_encoded = pd.get_dummies(df, columns=colunas_categoricas)

X = df_encoded.values

print(df.head())

X = df.iloc[:, :-1].values 
y = df.iloc[:, -1].values   

X_numerico = df_encoded[colunas_numericas]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numerico)

imputer = SimpleImputer(strategy='mean')
X_scaled_imputed = imputer.fit_transform(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_imputed)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='viridis', s=100, edgecolor='black')
plt.title('PCA: Visualização dos Grupos Identificados (2D)', fontsize=16)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Classe', loc='upper right')
plt.show()

kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(X_scaled_imputed)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_kmeans, palette='Set1', s=100, edgecolor='black', marker='o')
plt.title('PCA com Clustering KMeans', fontsize=16)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster', loc='upper right')
plt.show()

print("Componentes principais (cargas):")
print(pd.DataFrame(pca.components_, columns=X_numerico.columns, index=['PC1', 'PC2']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm


data = pd.read_csv(base)

print("\nVerificação de valores nulos nas colunas selecionadas:")
print(data[['Base armor', 'Base mana', 'Attack range', 'Attack damage']].isnull().sum())


X = data[['Base armor', 'Base mana', 'Attack range', 'Attack damage']]  
y = data['Base HP']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit() 

print("\nResumo do Modelo de Regressão Linear:")
print(model.summary())

print(f"\nCoeficientes do modelo: {model.params}")

plt.figure(figsize=(10, 6))

for var in X.columns[1:]:
    plt.scatter(data[var], y, label=f'{var} vs Base HP', alpha=0.6)
    plt.plot(data[var], model.fittedvalues, label=f'{var} Ajustado', color='red', linestyle='--')

plt.title('Relação entre Variáveis Preditivas e Base HP')
plt.xlabel('Variáveis Preditivas')
plt.ylabel('Base HP')
plt.legend()
plt.show()
```
