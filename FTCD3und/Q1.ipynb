```python
import pandas as pd

# Carregando o arquivo CSV
base = "champions.mod.csv"  # Substitua pelo caminho do seu arquivo
df = pd.read_csv(base)

# Selecionar apenas colunas numéricas
numerical_data = df.select_dtypes(include=['float64', 'int64'])

# Função para calcular estatísticas básicas
def calcular_estatisticas(df):
    estatisticas = {}
    for coluna in df.columns:
        valores = df[coluna].dropna()  # Remover valores nulos antes de todos os calculos
        estatisticas[coluna] = {
            'Média': valores.mean(),
            'Moda': valores.mode().tolist(), 
            'Mediana': valores.median(),
            'Variância': valores.var(),
            'Amplitude': valores.max() - valores.min(),
            'Mínimo': valores.min(),
            'Máximo': valores.max(),
            'Valores Distintos': valores.nunique()
        }
    return pd.DataFrame(estatisticas)

estatisticas_df = calcular_estatisticas(numerical_data)
print(estatisticas_df.T)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Carregar o arquivo CSV
base = "champions.mod.csv"  # Substitua pelo caminho do seu arquivo
df = pd.read_csv(base)

# Selecionar apenas colunas numéricas
numerical_data = df.select_dtypes(include=['float64', 'int64'])

# Análise de distribuição
for coluna in numerical_data.columns:
    valores = numerical_data[coluna].dropna()

    # Plotar Histograma e KDE
    plt.figure(figsize=(12, 6))
    sns.histplot(valores, kde=True, stat="density", bins=30, color='skyblue', label='Histograma')
    plt.title(f"Distribuição de {coluna}")
    plt.xlabel(coluna)
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

    # Gráfico Q-Q
    plt.figure(figsize=(8, 8))
    stats.probplot(valores, dist="norm", plot=plt)
    plt.title(f"Gráfico Q-Q de {coluna}")
    plt.show()

    # Teste de Shapiro-Wilk para normalidade
    stat, p_value = stats.shapiro(valores)
    print(f"Shapiro-Wilk Test para {coluna}: Estatística={stat:.4f}, p-valor={p_value:.4f}")
    if p_value > 0.05:
        print(f"-> {coluna} segue uma distribuição normal (p-valor > 0.05)\n")
    else:
        print(f"-> {coluna} NÃO segue uma distribuição normal (p-valor <= 0.05)\n")
```

```python
valores_faltosos = df.isnull().sum()

# Exibir resultado para todas as variáveis
print(valores_faltosos.sort_values(ascending=False))
```

```python
from sklearn.model_selection import train_test_split

test_data = df[df.isnull().any(axis=1)].copy()

data_without_missing = df.dropna()

train_data, val_data = train_test_split(data_without_missing, test_size=0.2, random_state=42) # 80% treino e 20% validação

print("Tamanho da base de treino:", len(train_data))
print("Tamanho da base de validação:", len(val_data))
print("Tamanho da base de teste:", len(test_data))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

cols_with_missing = df.columns[df.isnull().any()]

for col in cols_with_missing:
    X_train = train_data.drop(col, axis=1)
    y_train = train_data[col]
    X_val = val_data.drop(col, axis=1)
    y_val = val_data[col]

    knn_model = KNeighborsRegressor(n_neighbors=5)
    knn_model.fit(X_train, y_train)
    y_pred = knn_model.predict(X_val)

    mse = mean_squared_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)

    if mse < 0.1:
        print(f"{mse:.3f}: O valor do MSE é considerado bom, pois indica um baixo erro quadrático médio entre valores preditos e reais.")
    else:
        print(f"{mse:.3f}: O valor do MSE não é considerado tão bom, pode haver um erro quadrático médio entre valores preditos e reais considerável.")

    if r2 > 0.7:
        print(f"{r2:.3f}O valor do R² é considerado bom, pois indica uma alta correlação entre os valores preditos e reais, ou seja, o modelo consegue explicar uma grande parte da variância dos dados.")
    else:
        print(f"{r2:.3f}O valor do R² não é considerado tão bom, o modelo pode não estar capturando bem a relação entre as variáveis, sendo necessário uma melhor avaliação.")
    print("-" * 20)
```
